The target_class column is the one of interest. A 1 means a confirmed pulsar. Taken from the Kaggle description:

"*Each candidate is described by 8 continuous variables, and a single class variable. The first four are simple statistics obtained from the integrated pulse profile (folded profile). This is an array of continuous variables that describe a longitude-resolved version of the signal that has been averaged in both time and frequency . The remaining four variables are similarly obtained from the DM-SNR curve.*" 

A bit technical. But essentially a pulsar is a star that is spinning magnetized neutron star. It emits a beam of electromagnetic radiation and can be observed. The orientation, frequency of rotation, and several other specific characteristics are of both astrophysical and cosmological interest. Look up wikipedia, basically!

So the fascinating thing about pulsars is that their spin frequencies tell us things about the universe. This dataset appears to examine two key measurements: the '*integrated profile*' (IP for simplicity) and the '*DM-SNR*' curve.

I've never studied pulsars, but the former appears to be a measure of the 'regularity' of the characteristic rotation of each pulsar - from a quick google. Here is a typical example of an IP plotted:

<img src="IP1.png" alt="Drawing" style="width: 300px;"/>

Individual pulses seem to be somewhat stochastic, but the integrated - or summed - pulses over time appear to be quite regular and unique to the star.

So now we can conceptualise these IPs statistically, and get a feel for the standard deviation, excess kurtosis and skewness of the profile. It is probably a poor reference, but just for flavour the following are the properties of a standard univariate normal distribution: 

- 1 standard deviation =  68
- kurtosis (tail distribution) = 3
- skewness = 0

The latter, *DM-SNR*, is the 'Dispersion Measure Signal to Noise Ratio'. This measures the time dispersion of signals arriving at the detector due to atmospheric effects, interstellar dust, and other factors basically - I think! Then there is some SNR involved in the detection and the same measurements are applied, as above.

Similarly, it is fair to say there is a difference, at least visually! We should probably look at this a bit more numerically and see if we can quantify this a bit. First, as usual, we seperate into test and train.

```python
sample = np.random.rand(len(pulsar)) < 0.75 # take 75% of the pulsar set as training, 25% of test.
train = pulsar[sample]
test = pulsar[~sample]
```

```python
# Group the mean of each column by pulsar/not pulsar
mean_compare = pulsar.groupby("Class")[['Mean_IP','SD_IP','Kurt_IP','Skew_IP','Mean_DM','SD_DM','Kurt_DM','Skew_DM']].mean().reset_index()
# Have a quick look
mean_compare.head()
```

```python
# Group the std of each column by pulsar/not pulsar also
std_compare = pulsar.groupby("Class")[['Mean_IP','SD_IP','Kurt_IP','Skew_IP','Mean_DM','SD_DM','Kurt_DM','Skew_DM']].std().reset_index()
# Have a quick look
std_compare.head()
```

```python
mu_0 = mean_compare.iloc[0,1]
sigma_0 = std_compare.iloc[0,1] # Retrieve the mean and std for the non-pulsars

mu_1 = mean_compare.iloc[1,1]
sigma_1 = std_compare.iloc[1,1] # Retrieve the mean and std for the pulsars

x = np.linspace(mu_0 - 3*sigma_0, mu_0 + 3*sigma_0, 100)
y = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)

plt.plot(x, stats.norm.pdf(x, mu_0, sigma_0),label = 'Non-pulsars')
plt.plot(y, stats.norm.pdf(y, mu_1, sigma_0),label = 'Pulsars')
plt.legend(loc='upper left')
plt.fill_between(x,stats.norm.pdf(x, mu_0, sigma_0))
plt.fill_between(y,stats.norm.pdf(y, mu_1, sigma_0))
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# For the Integrated Profile:
mu_0 = mean_compare.iloc[0,1]
sigma_0 = std_compare.iloc[0,1] # Retrieve the mean and std for the non-pulsars

mu_1 = mean_compare.iloc[1,1]
sigma_1 = std_compare.iloc[1,1] # Retrieve the mean and std for the pulsars

x = np.linspace(mu_0 - 3*sigma_0, mu_0 + 3*sigma_0, 100)
y = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)

ax1.plot(x, stats.norm.pdf(x, mu_0, sigma_0),label = 'Non-pulsars')
ax1.plot(y, stats.norm.pdf(y, mu_1, sigma_0),label = 'Pulsars')
ax1.legend(loc='upper left')
ax1.fill_between(x,stats.norm.pdf(x, mu_0, sigma_0))
ax1.fill_between(y,stats.norm.pdf(y, mu_1, sigma_0))
ax1.set_ylabel('Integrated Profile')

# Now for the Dispersion Measure SNR:
mu_0a = mean_compare.iloc[0,5]
sigma_0a = std_compare.iloc[0,5] # Retrieve the mean and std for the non-pulsars

mu_1a = mean_compare.iloc[1,5]
sigma_1a = std_compare.iloc[1,5] # Retrieve the mean and std for the pulsars

xa = np.linspace(mu_0a - 3*sigma_0a, mu_0a + 3*sigma_0a, 100)
ya = np.linspace(mu_1a - 3*sigma_1a, mu_1a + 3*sigma_1a, 100)

ax2.plot(xa, stats.norm.pdf(xa, mu_0a, sigma_0a),label = 'Non-pulsars')
ax2.plot(ya, stats.norm.pdf(ya, mu_1a, sigma_0a),label = 'Pulsars')
ax2.legend(loc='upper left')
ax2.fill_between(xa,stats.norm.pdf(xa, mu_0a, sigma_0a))
ax2.fill_between(ya,stats.norm.pdf(ya, mu_1a, sigma_0a))
ax2.set_ylabel('Dispersion Measure - SNR')
```

Perhaps we can tidy up this code and run for all variables programatically. We can use the method above and put it into a loop.

```python
fig, (ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(8, 1, sharex=True)
fig.set_figheight(15)
fig.set_figwidth(15)

placeholder = [0] * 8    

for i in range (0,8):
    placeholder[i] = i
    plot_ax = 'ax' + str(placeholder[i])
    
    mu_0 = mean_compare.iloc[0,i+1]
    sigma_0 = std_compare.iloc[0,i+1] # Retrieve the mean and std for the non-pulsars

    mu_1 = mean_compare.iloc[1,i+1]
    sigma_1 = std_compare.iloc[1,i+1] # Retrieve the mean and std for the pulsars
    
    x = np.linspace(mu_0 - 3*sigma_0, mu_0 + 3*sigma_0, 100) # x and y values
    y = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)

    # Now plot!
    eval(plot_ax).plot(x, stats.norm.pdf(x, mu_0, sigma_0),label = 'Non-pulsars') 
    eval(plot_ax).plot(y, stats.norm.pdf(y, mu_1, sigma_0),label = 'Pulsars')
    eval(plot_ax).legend(loc='upper left')
    eval(plot_ax).fill_between(x,stats.norm.pdf(x, mu_0, sigma_0))
    eval(plot_ax).fill_between(y,stats.norm.pdf(y, mu_1, sigma_0))
    eval(plot_ax).set_ylabel(list(std_compare)[i+1])
```

# Some Machine Learning Models

We can start with the obvious choice - a logistic regression and look at some of its characteristics.

### 1. Logistic Regression

```python
# How many pulsars vs non pulsars in the data?
from __future__ import division # To stop float errors producing zero results
count_non_pulsar = len(pulsar[pulsar['Class']==0])
count_pulsar = len(pulsar[pulsar['Class']==1])
pct_of_non_pulsar = count_non_pulsar/(count_non_pulsar+count_pulsar)
print("Percentage of non-pulsars is: ", pct_of_non_pulsar * 100)
pct_of_pulsar = count_pulsar/(count_non_pulsar+count_pulsar)
print("Percentage of pulsars is: ", pct_of_pulsar * 100)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(pulsar.iloc[:,8],pulsar.iloc[:,0:7]) # The Class column is the variable of interest (column 8)
result=logit_model.fit()
print(result.summary2())
```

All variables appear to be statistically significant, with p-values of 0 (??) for all but Mean_IP. Taking a glance at the upper/lower bounds of the 95% confidence interval, we check that 0 is not contained - otherwise dubious inferences! 0 isn't contained here, so we may proceed. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

logreg = LogisticRegression()
logreg.fit(train.iloc[:,0:7], train.iloc[:,8]) # Build the model
y_pred = logreg.predict(test.iloc[:,0:7])
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(test.iloc[:,0:7], test.iloc[:,8])))
```

This is extremely high, but with high statistical significances for every variable, we may have expected accuracy of this order. Lets examine in further depth.

```python
# Confusion matrices show us true positives, false positives, false negatives and true negatives.
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(test.iloc[:,8], y_pred)
print(confusion_matrix)
```

Two interesting and standard measures of accuracy from the confusion matrix are *specificity* and *sensitivity*. The former is the true negative rate (how often the classifier correctly predicts a negative) and the latter is the true positive rate. A simple calculation shows:

```python
specificity = confusion_matrix[1][1] / (confusion_matrix[1][1] + confusion_matrix[1][0])
sensitivity = confusion_matrix[0][0] / (confusion_matrix[0][0] + confusion_matrix[0][1])
print("Sensitivity (TPR): ", sensitivity * 100)
print("Specificity (FPR): ", specificity * 100)
```

So with this model, we correctly identify a pulsar 99.38% of the time, whereas we correctly identify a non-pulsar 84.46% of the time. To be continued!
